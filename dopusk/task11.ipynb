```python
class MealyError(Exception):
    pass


brake = {'A': 'B', 'E': 'E'}
rev = {'B': 'B', 'C': 'D', 'D': 'E', 'E': 'B'}
dash = {'B': 'C', 'E': 'F', 'F': 'D'}
brake1 = {'A': 0, 'E': 7}
rev1 = {'B': 2, 'C': 3, 'D': 4, 'E': 6}
dash1 = {'B': 1, 'E': 5, 'F': 8}


class Mealy:
    def __init__(self):
        self.state = 'A'

    def brake(self):
        if self.state in brake:
            a = self.state
            self.state = brake[self.state]
            return brake1[a]
        else:
            raise MealyError('brake')

    def rev(self):
        if self.state in rev:
            a = self.state
            self.state = rev[self.state]
            return rev1[a]
        else:
            raise MealyError('rev')

    def dash(self):
        if self.state in dash:
            a = self.state
            self.state = dash[self.state]
            return dash1[a]
        else:
            raise MealyError('dash')


def test():
    mealy = main()
    try:
        mealy.rev()
    except MealyError:
        pass
    try:
        mealy.dash()
    except MealyError:
        pass
    mealy.brake()
    mealy.rev()
    try:
        mealy.brake()
    except MealyError:
        pass
    mealy.dash()
    mealy.rev()
    mealy.rev()
    mealy.brake()
    mealy.dash()
    mealy.dash()
    mealy.rev()
    mealy.rev()


def main():
    return Mealy()
```
