```python
class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


def sec(x, x0, x1, x2, x3, x4):
    if x0.dataval == 1997:
        if x3.dataval in (1964, 1957):
            x0.nextval = 9 + (1964 != x3.dataval)
        else:
            x0.nextval = 7 + (1974 != x1.dataval)
        return x0, x1, x2, x3, x4
    x0.nextval = x2 if x1.dataval == 1974 else x3
    if x3.dataval in (1964, 1957):
        x3.nextval = 4 + (1964 != x3.dataval)
    else:
        x3.nextval = 3
    return x0, x1, x2, x3, x4


def calc(x, x0, x1, x2, x3, x4):
    x4.nextval = x0 if x4.dataval == 'C' else 11
    if x0.dataval == 1999:
        x0.nextval = 6
    if x2.dataval in ('ALLOY', 'M4'):
        x2.nextval = 0 + ('ALLOY' != x2.dataval)
    else:
        x2.nextval = 2
    return x0, x1, x2, x3, x4


def main(x):
    x4 = Node()
    x3 = Node()
    x2 = Node()
    x1 = Node()
    x0 = Node()
    x4.dataval = (x[4])
    x3.dataval = (x[3])
    x2.dataval = (x[2])
    x1.dataval = (x[1])
    x0.dataval = (x[0])
    x0, x1, x2, x3, x4 = sec(x, x0, x1, x2, x3, x4)
    x0, x1, x2, x3, x4 = calc(x, x0, x1, x2, x3, x4)
    obj = x4
    while type(obj.nextval) != int:
        obj = obj.nextval
    return obj.nextval
```
